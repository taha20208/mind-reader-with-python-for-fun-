```python
import tkinter as tk
from tkinter import ttk


entry = None
result_label = None


progress_texts = ["Progressing...", "I'm going to find your think", "I almost find it!"]
current_progress_text = [0]  

def show_think():
    global entry, result_label  
    think = entry.get()

    
    progress_window = tk.Toplevel(root)
    progress_window.title("Progress")
    progress_window.geometry("300x100")

    
    progress = ttk.Progressbar(progress_window, length=250, mode="indeterminate")
    progress.pack(pady=10)


    progress_label = tk.Label(progress_window, text=progress_texts[current_progress_text[0]])
    progress_label.pack()

   
    progress["value"] = 20
    progress_window.update_idletasks()


    def update_progress_text():
        current_progress_text[0] += 1
        if current_progress_text[0] < len(progress_texts):
            progress_label.config(text=progress_texts[current_progress_text[0]])
            progress_window.after(5000, update_progress_text)  
        else:
            progress.stop()
            progress.destroy()
            progress_label.destroy()
            update_progress(think, progress_window)

    
    progress.after(400, update_progress_text)
    progress.start(20)

def update_progress(think, window):
    
    window.destroy()

    
    result_window = tk.Toplevel(root)
    result_window.title("Result")

    
    result_label = tk.Label(result_window, text=f'You are thinking of: {think} ðŸ’€' , fg="red")
    result_label.pack()

def main():
    global entry, result_label, root  
    root = tk.Tk()
    root.title("mind reader")

    entry_label = tk.Label(root, text="Please enter your think:")
    entry_label.pack()

    entry = tk.Entry(root)
    entry.pack()

    enter_button = tk.Button(root, text="Enter", command=show_think)
    enter_button.pack()

    result_label = tk.Label(root, text="")
    result_label.pack()

    root.mainloop()

if __name__ == "__main__":
    main()
```
